## Tranformação de variáveis : LabelEncoder e OneHotEncoder

Assim como na estatística, as informações na área da ciência de dados podem ser divididas em duas classes: Núméricos e Categóricos.
Em cada uma das classes existem subdivisões atreladas ao sentido de suas informações. A classe numérica se divide em dados dicretos(números inteiros) e contínuos(números reais), por sua vez, a classe categórica se divide em dados ordinais e nominais.

<img src="img/dados.png" width=580 height=500 />


Ao utilizarmos algoritimos de machine learning, devemos nos atentar para os tipos de dados das variaveis preditivas e classificatorias e assegurar a tranformação dessas variáreis para o tipo numperico, para isso a bilioteca sklearning consta com dois métodos



```python
import pandas as pd
data = {'Empregador':['Corpo de Bombeiros', 'Secretaria de Saúde', 'Polícia Civil', 'Secretaria de Saúde', 'Polícia Civil'],
       'Idade': [44,34,46,35,23], 'Remuneração':[72000,65000,98000,45000,34000]}
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

df['Empregador'] = label_encoder.fit_transform(df['Empregador'])
print(df)
```

```python
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder()

#remodela a matriz de 1-D para 2_D, ja que fit_transform espera 2_D e posteriormente ajusta o objeto
x = onehotencoder.fit_transform(df.Empregador.values.reshape(-1,1)).toarray()
x
```

```python
# para adicionar de volta no dataframe original
dfOneHot = pd.DataFrame(x, columns = ['Empregador_'+str(int(i)) for i in range(df.shape[1])])
df = pd.concat([df, dfOneHot], axis=1)

# dropando a coluna original
df = df.drop(["Empregador"], axis=1)

print(df)
```
